```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('C:\\Users\\mahik\\OneDrive\\Desktop\\datasetrestro\\Dataset .csv')
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['Longitude'], data['Latitude'], alpha=0.5)
plt.title('Distribution of Restaurants')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```

```python
restaurants_by_locality = data.groupby('Locality').size().reset_index(name='restaurant_count')
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Locality', y='restaurant_count', data=restaurants_by_locality.head(20))
plt.title('Top 20 Localities with Most Restaurants')
plt.xlabel('Locality')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=90)
plt.show()
```

```python
avg_ratings_by_locality = data.groupby('Locality')['Aggregate rating'].mean().reset_index()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Locality', y='Aggregate rating', data=avg_ratings_by_locality.head(20))
plt.title('Average Ratings by Locality')
plt.xlabel('Locality')
plt.ylabel('Average Rating')
plt.xticks(rotation=90)
plt.ylim(0, 5)  # Set y-axis limit to 0-5 for ratings
plt.show()
```

```python
avg_price_by_locality = data.groupby('Locality')['Price range'].mean().reset_index()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Locality', y='Price range', data=avg_price_by_locality.head(20))
plt.title('Average Price Range by Locality')
plt.xlabel('Locality')
plt.ylabel('Average Price Range')
plt.xticks(rotation=90)
plt.show()
```

```python
cuisine_count_by_locality = data.groupby('Locality')['Cuisines'].nunique().reset_index(name='cuisine_count')
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Locality', y='cuisine_count', data=cuisine_count_by_locality.head(20))
plt.title('Cuisine Diversity by Locality')
plt.xlabel('Locality')
plt.ylabel('Number of Unique Cuisines')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Price range', y='Aggregate rating', data=data)
plt.title('Correlation between Ratings and Price Range')
plt.xlabel('Price Range')
plt.ylabel('Aggregate Rating')
plt.show()
```
